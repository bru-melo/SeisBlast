```python
#Create input file for seismic tomography --- from the uniq quarry-station file
import pandas as pd
#import SeisBlast as sb
seisblast_files = '/home/bmelo/bin/SeisBlast/files/'
```

```python
#Import all the necessary files
#james_catalogue = pd.read_csv('/mnt/REPO/QUARRY-BLASTS/CATALOGS/quarry-blast-catalog-2013-2014.csv', sep = r'\s+')
#james_bulletins_dir = "/mnt/REPO/QUARRY-BLASTS/BULLETINS/"
stations = pd.read_csv(
    "/home/bmelo/bin/Quarry/James_files/Total_Stations_List.txt", 
    delimiter=' ',
    header = 0,
    names=['sta', 'name','lat', 'lon', 'elevation', 'net', 'end'])
uniq_blasts = pd.read_csv('/home/bmelo/bin/SeisBlast/files/uniq_quarry-station_list.csv')
uniq_blasts
#merge with the relocated quarry blasts lists to get the relocated events
#relocated_catalogue = pd.merge(james_catalogue, relocated_blasts, left_on='EVENT-ID', right_on='id-ev')
```

```python

#start with formating the separate date and time formats
uniq_blasts['year'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.year
uniq_blasts['month'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.month
uniq_blasts['day'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.day
uniq_blasts['hour'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.hour
uniq_blasts['minute'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.minute
uniq_blasts['second'] = pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.second + pd.to_datetime(uniq_blasts['origin_time'], format='%Y-%m-%d %H:%M:%S.%f').dt.microsecond / 1e6
#uniq_blasts['depth-quarry'] = 0
```

```python
#Source line : id_src year month day hour min sec lat lon dep_km magnitude num_recs id_event (weight)
#Currently building it up with only P phases !!!!

# Open the file in write mode to ensure it's empty before starting the loop
with open(seisblast_files + '/seismic_tomography_uniqinput_P.txt', 'w') as f:
    f.write("")
    
for index_src in range(len(uniq_blasts)):
    
    #loop over the rows in the dataframe
    len_rec = len(eval(uniq_blasts['sta'].iloc[index_src]))
    #print(str(uniq_blasts['count'].iloc[index_src]))
    with open(seisblast_files + '/seismic_tomography_uniqinput_P.txt', 'a') as f:
        f.write(
            f"{index_src:<8}"
            f"{uniq_blasts['year'].iloc[index_src]:<8}"
            f"{uniq_blasts['month'].iloc[index_src]:<8}"
            f"{uniq_blasts['day'].iloc[index_src]:<8}"
            f"{uniq_blasts['hour'].iloc[index_src]:<8}"
            f"{uniq_blasts['minute'].iloc[index_src]:<8}"
            f"{uniq_blasts['second'].iloc[index_src]:<12.2f}"
            f"{uniq_blasts['quarry-lat'].iloc[index_src]:<12}"
            f"{uniq_blasts['quarry-lon'].iloc[index_src]:<12}"
            f"{'0':<12}"
            f"{uniq_blasts['mag'].iloc[index_src]:<12}"
            f"{str(len_rec):<12}"
            f"{uniq_blasts['closest-ev_id'].iloc[index_src]:<12}\n"
        )
        
        #loop over the items in each row
        for index_rec in range(len_rec):
            #fetch station information
            st=eval(uniq_blasts['sta'].iloc[index_src])[index_rec]
            station_info = stations[stations['name'] == st]
            #write the receiver lines to file
            f.write(
                f"{index_src:<8}"
                f"{index_rec:<8}"
                f"{st:<10}"
                f"{station_info['lat'].item():<12}"
                f"{station_info['lon'].item():<12}"
                f"{station_info['elevation'].item():<12}"
                f"{'P':<8}"
                f"{eval(uniq_blasts['closest-time'].iloc[index_src])[index_rec]:<12}\n"
            )
```
